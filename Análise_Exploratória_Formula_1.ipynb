# Analisando a nacionalidade dos pilotos

```python
# Importando pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando o dataset dos pilotos
pilotos = pd.read_csv('drivers.csv')
```

```python
# Visualizar o dataset pilotos
pilotos.head()
```

```python
# Formato da tabela
pilotos.shape
```

```python
#Nacionalidade dos pilotos
nacionalidades_pilotos = pilotos['nationality'].unique()
nacionalidades_pilotos
```

```python
# Contagem da nacionalidade dos pilotos
contagem_nacionalidade_pilotos = pilotos['nationality'].value_counts()
contagem_nacionalidade_pilotos
```

```python
# Criando um dataframe com a contagem das nacionalidades dos pilotos
contagem_nacionalidade_pilotos = contagem_nacionalidade_pilotos.to_frame().reset_index()
contagem_nacionalidade_pilotos.columns = ['nationality', 'count']
contagem_nacionalidade_pilotos
```

```python
# um barplot com a contagem da nacionalidade dos pilotos (países com mais de 20 pilotos)
fig = plt.figure(figsize=(12,4))
sns.barplot(data = contagem_nacionalidade_pilotos.query('count >= 20'), x='nationality', y='count', palette='BuGn_d')
plt.xlabel('Nacionalidade')
plt.ylabel('Quantidade')
plt.title('Quantidade de pilotos na F1 por nacionalidade')
plt.show()
```

# Analisando a média das posições dos pilotos

```python
# Carregando o arquivo dos resultados dos pilotos
resultados_pilotos = pd.read_csv('results.csv')
```

```python
# Visualizando a tabela
resultados_pilotos.head()
```

```python
# Ver os tipo de variáveis numa tabela
resultados_pilotos['number'].dtypes.name
```

```python
# Excluindo as colunas que possui tipo de variável object
excluir = []
for col in resultados_pilotos.columns:
  if resultados_pilotos[col].dtypes.name == 'object':
    excluir.append(col)

resultados_pilotos = resultados_pilotos.drop(columns=excluir)
```

```python
resultados_pilotos.head()
```

```python
resultados_pilotos.shape
```

```python
#criado uma lista para colocar como coluna no DataFrame
sobrenomes = []
for pilotoId in resultados_pilotos.driverId:
  texto = 'driverId==' + str(pilotoId)
  nome = list(pilotos.query(texto)['surname'])[0]
  sobrenomes.append(nome)
```

```python
# Adicionando uma coluna como DataFrame
resultados_pilotos.insert(2, 'driver_surname', sobrenomes)
```

```python
resultados_pilotos
```

```python
sns.boxplot(data=resultados_pilotos.query('driver_surname in ["Hamilton", "Alonso", "Senna", "Barrichello"]'), x='driver_surname', y='positionOrder')
```
